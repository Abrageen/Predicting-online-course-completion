# Importing libraries


```python
from sklearn import datasets 
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

# Reading data

```python
db=pd.read_csv(r"C:\Users\Divyansh Gupta\Desktop\Datasets\Trend Data.csv")
print(db.head())
```

# Splitting data into features and labels

```python
x1=db['Days passed']
y1=db['Completion Status_t']
x = np.array(x1)
y = np.array(y1)
x=x.reshape(-1,1)
y=y.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
print(y.ndim)
```

# Training the random forest model

```python
clf = RandomForestClassifier(n_estimators = 100)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))
```

# Plotting one of the decision trees used in the model

```python
fig = plt.figure(figsize=(15, 10))
db['Completion Status_t'] = pd.Series(db['Completion Status_t'])
db['Days passed'] = pd.Series(db['Days passed'])
plot_tree(clf.estimators_[21]) 
plt.show()
```

# Predicting probability

```python
e=[35]
e = np.array(e)
e=e.reshape(-1,1)
print(clf.predict(e))
predictions = clf.predict_proba(e)
print(predictions)
```
